# What is Data Governance?

Cross functional discipline of managing, imporving, monitoring, maintaining, and protecting data. 

Basically, Proactively managing data to support business achieve its strategy and vision.

(or) Simply:
A data management function to ensure the quality, integrity, security, and unsability of the data collected.

(or) higher level
Data well managed!

# Priciples of DG

1. Transparency 
    - Avoid surprises
    - Gain buy-in
    - Build trust
    - Understand why things are being done 
    - Understand the valkue of those actions
    
    Eg: Organization may have an information security policy that clearly states what kind of data must never be displayed on
    its public website. The policy should state why this is important and how it's being enforced.
2. Accountability
    Resposibilities of each an every individually within the organization
    - Fully described
    - Understood
    - Agreed upon by all parties
    - Held accountable for taking data-related actions at specific times
    
    Eg: If a role requires a person to disclose what stocks they hold and that they must enter that data by a certain date,
    by not adhering to this requirement could result in termination of employment.
3. Standardization
    We are concerned with such things as how data is consistently labeled, described and categorized. 
    - Improes quality of data
    - Helps organizational communication
    - Makes data more usable across systems
    - Makes data more valuable(overall result)
    
    Eg: How an organization might handle US state designations. When California is being entered in a system, is the 
    standard to use CA or to spell out the entire state name or something else?

Reference data or Master data management:
    To increase the ability for data to be governed, understood by stakeholders and even shared across enterprise systems, there must be organizational agreement on the details of reference data. By reference data we can think of core data sets that help tun the organization day-to-day. Data such as customer and supplier details and product and part attributes. Reference data is better known as master data in the data science space. Master data management uses this standardization to enable better enterprise operations and data integration within the organization and with external entities. Data governance is not limited to these principles, but they represent the type of discipline and attention that is required.

# Risk Management
    Minimize risk exposure to sensitive information to unauthorized individuals/systems, security breaches or
    known personnel accessing data under wrong circumstances. This can be achieved by encryptions (2FA), access management, 
    identity sensitive data assests and create a policy around protection.
# Data Proliferation
    The speed at which businesses create, update, and stream their data assets has increased, and while
    cloud based platforms are capable of handling increased data velocity, volume, and variety, it is important to 
    introduce controls and mechanisms to rapidly validate the quality ascpects of high-bandwisth data streams
# Data Management
    The need to adopt externally produced data sources and data streams means that you should be prepared not 
    to trust all external data sources. you may need to introduce tools that document data lineage, classification, and 
    metadata to help employees(data consumers) to determine data usability based on their knowledge of how the data assets 
    were produced.

# Discovery(and data awareness)
    Moving data from a data lake or to a data warehourse or a cloud-based system, has a risk of losing track of which data 
    assets have been moved, the characteristics of their contect and details about their metadata. The ability to assess data 
    asset content and sensitivity becomes very important.

# Best Practices to adapt

1. Start small. As in all aspects of business, do not try to boil the ocean. 
Strive for quick wins and build up ambitions over time.

2. Set clear, measurable, and specific goals. You cannot control what you cannot measure. 
Celebrate when goals are met and use this to go for the next win.

3. Define ownership. Without business ownership a data governance framework cannot succeed.

4. Identify related roles and responsibilities. Data governance is a teamwork with deliverables from all parts of the business.

5. Educate stakeholders. Wherever possible use business terms and translate the academic parts of the data governance
discipline into meaningful content in the business context.

6. Focus on the operating model. A data governance framework must integrate into the way of doing business in your enterprise.

7. Map infrastructure, architecture, and tools. Your data governance framework must be a sensible part of your enterprise 
architecture, the IT landscape and the tools needed.

8. Develop standardized data definitions. It is essential to strike a balance between what needs to be centralized and 
where agility and localization works best.

9. Identify data domains. Start with the data domain with the best ratio between impact and effort for rising the 
data governance maturity.

10. Identify critical data elements. Focus on the most critical data elements.

11. Define control measurements. Deploy these in business process, IT applications and/or reporting where it makes most sense.

12. Build a business case. Identify advantages of rising data governance maturity related to growth, costs savings, risk 
and compliance.

13. Leverage metrics. Focus on a limited set of data quality KPIs that can be related to general performance KPIs within 
the enterprise.

14. Communicate frequently. Data governance practitioners agree that communication is the most 
crucial part of the discipline.

15. It’s a practice, not a project.
