```python
import matplotlib.pyplot as mp
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
uniform_data = np.random.rand(10,12)
```

```python
data = pd.DataFrame({'x':np.arange(1,101), 'y':np.random.normal(0,4,100)})
```

```python
# Build in dataset of seaborn 
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

```python
# Figure aesthetics
f, ax = mp.subplots(figsize = (5,6))
```

```python
# Seaborn styling
sns.set()
sns.set_style("whitegrid")
sns.set_style("ticks", {"xtick.major.size":8, "ytick.major.size":8})
sns.axes_style("whitegrid")
```

```python
# Context function
sns.set_context("talk") # contect to "talk"
sns.set("notebook", font_scale = 1.5, rc = {"lines.linewidth":2.5}) # contect to "notebook", scale font elements and override param mapping
```

```python
# color palette
sns.set_palette("husl",3) # define the color palette
sns.color_palette("husl") # temporarily set palette
flatui = ["#9b59b6", "#3498bd", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.set_palette(flatui) # set any color palette
```

# Plotting with Searborn

```python
# Axis Grids
g = sns.FacetGrid(titanic,col="survived", row="sex") # subplot grid for plotting conditional relationships
g = g.map(mp.hist,"age")
```

```python
sns.factorplot(x = "pclass", y = "survived", hue = "sex", data=titanic) # draw a categorical plot onto a Facetgrid
```

```python
sns.lmplot(x="sepal_width",y="sepal_length",hue="species",data=iris) # plot data and regression model fits across a FacetGrid
```

# Categorical Plots

```python
# Scatterplot
sns.stripplot(x="species", y = "petal_length",data=iris) # Scatterplot with one categorical variable
```

```python
#Scatterplot
sns.swarmplot(x="species", y = "petal_length",data=iris) # Categorical scatterplot with non-overlapping points
```

```python
# Bar Chart
sns.barplot(x="sex", y ="survived",hue="class",data=titanic) #show points estimates and confidence intervals with scatterplot glyphs
```

```python
# Count Plot
sns.barplot(x="sex", y ="survived",hue="class",data=titanic) #Show count of observations
```

```python
# Point Plot
sns.pointplot(x="class", y="survived", hue="sex",data=titanic) # Show point estimates and confidence intervals as rectangular bars
```

```python
# Boxplot
sns.boxplot(x="alive", y="age", hue="adult_male",data=titanic)
```

```python
i = sns.JointGrid(x="x", y = "y", data=data)
i = i.plot(sns.regplot,sns.displot)
```

```python
sns.jointplot(x ="sepal_length",y ="sepal_width",data=iris,kind="kde")
```

# Regression Plots

```python
sns.regplot(x="sepal_width",y="sepal_length",data=iris) #Plot data and a linear regression model fit
```

# Distribution plots

```python
plot = sns.displot(data.y,kde=False,color="b") # Plot univariable distribution
```

# Matrix plots

```python
sns.heatmap(uniform_data,vmin=0, vmax=1) 
# Heatmaps are used to show relationships between two variables, one plotted on each axis. By observing how cell colors change across each axis, you can observe if there are any patterns in value for one or both variables
```

# Customizations

```python
#Axisgrid Objects
g.despine(left=True) # remove left spine
g.set_ylabels("Survived") #set the labels of the y-axis
g.set_xticklabels(rotation=45) #set the tick labels for x
g.set_axis_labels("Survived","Sex") #set the axis labels
h.set(xlim =(0,5),ylim=(0,5),xticks=[0,2.5,5],yticks=[0,2.5,5])
```

```python
#Plot x,y modifications
mp.title("Some_title")
mp.ylabel("Some_y_axis_label")
mp.xlabel("Some_x_axis_label")
mp.ylim(0,100)
mp.xlim(0,10)
mp.setp(ax,yticks=[0,5])
mp.tight_layout()

```

```python
mp.show() #show plot
mp.savefig("foo.png") #Save the plot as a figure
mp.savefig("foo.png",transparent=True) # Save transparent figure
```

```python
mp.cla() # clear an axis
mp.clf() # clear an entire figure
mp.close() #close a window
```
